```python
#Normal distribution.
def normalpdf(x, mu=0.0, sigma=1.0):
    """Calculates the PDF of a normal distrubtion."""
    numer = np.exp(-0.5 * np.square((x - mu) / sigma))
    denom = sigma * np.sqrt(2.0 * np.pi)
    return (numer / denom)
```

## Fill between

```python
# Create some x values.
x = np.linspace(-5.0, 5.0, 1000)

# Calculate our y values from x.
y = normalpdf(x)

# Create a plot.
fig, ax = plt.subplots()

# Plot the values (k= black).
ax.plot(x, y, 'k-')

# Limit x values to -2.0 and 2.0
x = np.linspace(-2.0, 2.0, 1000)

# Calculate the pdf values.
y = normalpdf(x)

# Fill between x axis and y - for x values between -2.0 and 2.0.
# fill area between two Y values i.e where y =0 i.e. the x axis and Y (the curve)
ax.fill_between(x, 0, y)

# Draw a couple of vertical lines.
# axvline draws vertical lines
ax.axvline(-2.0, color='grey', linestyle=':')
ax.axvline(2.0, color='grey', linestyle=':')

# Draw a horizontal line.

plt.axhline(y = 0.5, color = 'r', linestyle = 'dashed')

# Draw some text.
ax.text(0, 0.2, '~95%', horizontalalignment='center')
```

```python
# Extract sepal lengths.
# draws a normal pdf assuming each of the values as the mean
# change the number of vals values below to add/remove pdf's
vals = df['sepal_length'].to_numpy()[:10]

# Minimum and maximum values.
low, high = vals.min(), vals.max()

# Create some space on the x axis.
diff = high - low
low = low - diff
high = high + diff

# Create some x values.
# i.e. creating the values on the x axis with sufficient space eitehr side of curve
x = np.linspace(low, high, 1000)

# Create a plot.
fig, ax = plt.subplots()

#loop through the values in vals (sepal length)
# Select a value from the data.
for i in vals:
    # Calculate our y values from x.
    y = normalpdf(x, mu=i, sigma=0.100)
                  
# Plot the values.
    ax.plot(x, y, 'k-')
                  
# Fill between x axis and y.
    ax.fill_between(x, 0, y) 
```

#### creating the kde
- every time we calculate the Y value for each x value (i.e. peak of curve)
- rather than fill between the x axes fill bwtween the current pdf and previous ones

```python
# Extract sepal lengths.
vals = df['sepal_length'].to_numpy()[:50]
                                    

# Minimum and maximum values.
low, high = vals.min(), vals.max()

# Create some space on the x axis.
diff = high - low
low = low - diff
high = high + diff

# Create some x values.
# i.e. creating the values on the x axis with sufficient space eitehr side of curve
x = np.linspace(low, high, 1000)

# Create a plot.
fig, ax = plt.subplots()

# Set the initial previous y values to 0.
# (could be just set to 0 as 0*X = 0 but just for consistency here as x*0)
prev = 0 * x

#loop through the values in vals (sepal length)
# Select a value from the data.
for i in vals:
    # Calculate our y values from x.
    y = normalpdf(x, mu=i, sigma=0.1)
    # Add the previous sums of the y values.
    y = y + prev

    # Fill between x axis and y. Alpha = translucence
    ax.fill_between(x, prev, y, alpha=0.3)
    
    # Update the prev array.
    prev = y
    

# Plot the values.
ax.plot(x, y, 'k-')

# blue is 1st pdf
# yellow is 2nd additive and so on
#black line is kernel density estimate (KDE)
#note if StDev is 0.1.  If make it too large then too much overlap and no definition
# can see 1st 50 sepal lengths (setosas)
#reasonable question is what is the right value to set the sigma to
#kde can use other functions than normal (PDF).  Up to the data analyst to choose the right one
#selecting the bandwith (st dev) is similar to slection of bins in histogram
```
