# Overview of matplotlib.pyplot python package <br><br>  <img src="images/logo2.svg" width="200"/>

Python offers a variety of packages for plotting data.<br>
Matplotlib is a library for data visualization, typically in the form of plots, graphs and charts.<br>
**Pyplot** is an API (Application Programming Interface) for Python’s **matplotlib**. <br>
Python effectively makes matplotlib a viable open source alternative to **MATLAB**1. <br>
This Notebook is a guide to pyplot.<br><br><br>

- MATLAB® is a programming platform designed specifically for engineers and scientists to analyze and design systems and products that transform our world. The heart of MATLAB is the MATLAB language, a matrix-based language allowing the most natural expression of computational mathematics.
- Most of the Matplotlib utilities lies under the pyplot submodule and are usually imported under the plt alias: "matplotlib.pyplot as plt"

- matplotlib.pyplot is accessible via python if not already on your machine recommend to download python via downloading [Anaconda](https://www.anaconda.com/products/individual)
- To access the functionality of matplotlib.pyplot the following libraries should be downloaded:<br>

### Labels
#### use the xlabel() and ylabel() functions to set a label for the x- and y-axis.

```python
# plot x vs 2 x y arrays
# adding x & y labels
# adding titles
# adding gridlines
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.arange(1, 10, 2.5)
y2 = np.array([6, 2, 7, 11])


font1 = {'family':'serif','color':'blue','size':15}
font2 = {'family':'serif','color':'darkred','size':15}
# loc = left , right  or center
plt.plot(x1, y1, x2, y2)

# loc = left , right  or center

#plot the title
plt.title("Mfg line 1", loc = 'right')

#plot the x label. Use'fontdict= attributes contained in dict object as above
plt.xlabel("rpm", loc ='left', fontdict = font1)

#plot the y label.Use 'fontdict= attributes contained in dict object as above
plt.ylabel("Quantity", loc = 'top', fontdict = font2)

#plot a grid
#plt.grid()
#plt.grid(axis ='x')
plt.grid(axis ='y', color = 'green', linestyle = ':', linewidth = 2)
```

### Sub-plots
#### How to display more than one plot in one output

```python
# plotting side to side plots (horizontal)
#plot 1:
x = np.arange(1, 10, 2.5)
y = np.array([3, 8, 1, 10])

#SYNTAX: plt.subplot(ROWS, COLUMNS, SUBLPLOT #)
#define the location of the subplot 1
#the figure has 1 row, 2 columns, and this plot is the first plot
plt.subplot(1, 2, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

#define the location of the subplot 2
#the figure has 1 row, 2 columns, and this plot is the second plot.
plt.subplot(1, 2, 2)
plt.plot(x,y);
```

```python
# plotting side to side plots (top & bottom)

#plot 1:
x = np.arange(1, 10, 2.5)
y = np.array([3, 8, 1, 10])

#define the location of the subplot 1
#the figure has 2 row, 1 columns, and this plot is the first plot
plt.subplot(2, 1, 1)
plt.plot(x,y)

#define the location of the subplot 2
#the figure has 2 row, 1 columns, and this plot is the 2nd plot
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 1, 2)
plt.plot(x,y);
```

```python
# multiple plots (i.e. 6 in 2 rows)
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

#the figure has 2 row, 3 columns, i.e. 2 x 3 or 6 subplots & this is the first plot
plt.subplot(2, 3, 1)
plt.plot(x,y)

x = np.array([10, 20, 30, 40, 50])
y = np.array([10, 20, 30, 40, 50])

plt.subplot(2, 3, 2)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 3)
plt.plot(x,y)

x = np.arange(1, 10, 2.5)
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 4)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 5)
plt.plot(x,y)

x = np.arange(1, 100, 25)
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 6)
plt.plot(x,y);
```

```python
# colours cme font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

#plot the title
plt.title("Car watching", loc = 'center')

#plot the x label. Use'fontdict= attributes contained in dict object as above
plt.xlabel("Car age", loc ='center', fontdict = font1)

#plot the y label.Use 'fontdict= attributes contained in dict object as above
plt.ylabel("Car speed", loc = 'center', fontdict = font2)

#observations:
x = np.arange(1, 27, 2)
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
#individually colour the points based on a new array- quantity of array numbers/colors has to match
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])

plt.scatter(x, y, c=colors);
```

```python
# for umerical arrays
import numpy as np
# for plotting
import matplotlib.pyplot as plt
#for formatting i
from matplotlib.ticker import AutoMinorLocator, MultipleLocator, FuncFormatter


np.random.seed(19680801)
# create an array of random mumbers
X = np.linspace(0.5, 3.5, 100)
Y1 = 3+np.cos(X)
Y2 = 1+np.cos(1+X/0.75)/2
Y3 = np.random.uniform(Y1, Y2, len(X))

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(1, 1, 1, aspect=1)


def minor_tick(x, pos):
    if not x % 1.0:
        return ""
    return "%.2f" % x

ax.xaxis.set_major_locator(MultipleLocator(1.000))
ax.xaxis.set_minor_locator(AutoMinorLocator(4))
ax.yaxis.set_major_locator(MultipleLocator(1.000))
ax.yaxis.set_minor_locator(AutoMinorLocator(4))
ax.xaxis.set_minor_formatter(FuncFormatter(minor_tick))

ax.set_xlim(0, 4)
ax.set_ylim(0, 4)

ax.tick_params(which='major', width=1.0)
ax.tick_params(which='major', length=10)
ax.tick_params(which='minor', width=1.0, labelsize=10)
ax.tick_params(which='minor', length=5, labelsize=10, labelcolor='0.25')

ax.grid(linestyle="--", linewidth=0.5, color='.25', zorder=-10)

ax.plot(X, Y1, c=(0.25, 0.25, 1.00), lw=2, label="Blue signal", zorder=10)
ax.plot(X, Y2, c=(1.00, 0.25, 0.25), lw=2, label="Red signal")
ax.plot(X, Y3, linewidth=0,
        marker='o', markerfacecolor='w', markeredgecolor='k')

ax.set_title("Anatomy of a figure", fontsize=20, verticalalignment='bottom')
ax.set_xlabel("X axis label")
ax.set_ylabel("Y axis label")

ax.legend()


def circle(x, y, radius=0.15):
    from matplotlib.patches import Circle
    from matplotlib.patheffects import withStroke
    circle = Circle((x, y), radius, clip_on=False, zorder=10, linewidth=1,
                    edgecolor='black', facecolor=(0, 0, 0, .0125),
                    path_effects=[withStroke(linewidth=5, foreground='w')])
    ax.add_artist(circle)


def text(x, y, text):
    ax.text(x, y, text, backgroundcolor="white",
            ha='center', va='top', weight='bold', color='blue')


# Minor tick
circle(0.50, -0.10)
text(0.50, -0.32, "Minor tick label")

# Major tick
circle(-0.03, 4.00)
text(0.03, 3.80, "Major tick")

# Minor tick
circle(0.00, 3.50)
text(0.00, 3.30, "Minor tick")

# Major tick label
circle(-0.15, 3.00)
text(-0.15, 2.80, "Major tick label")

# X Label
circle(1.80, -0.27)
text(1.80, -0.45, "X axis label")

# Y Label
circle(-0.27, 1.80)
text(-0.27, 1.6, "Y axis label")

# Title
circle(1.60, 4.13)
text(1.60, 3.93, "Title")

# Blue plot
circle(1.75, 2.80)
text(1.75, 2.60, "Line\n(line plot)")

# Red plot
circle(1.20, 0.60)
text(1.20, 0.40, "Line\n(line plot)")

# Scatter plot
circle(3.20, 1.75)
text(3.20, 1.55, "Markers\n(scatter plot)")

# Grid
circle(3.00, 3.00)
text(3.00, 2.80, "Grid")

# Legend
circle(3.70, 3.80)
text(3.70, 3.60, "Legend")

# Axes
circle(0.5, 0.5)
text(0.5, 0.3, "Axes")

# Figure
circle(-0.3, 0.65)
text(-0.3, 0.45, "Figure")

color = 'blue'
ax.annotate('Spines', xy=(4.0, 0.35), xycoords='data',
            xytext=(3.3, 0.5), textcoords='data',
            weight='bold', color=color,
            arrowprops=dict(arrowstyle='->',
                            connectionstyle="arc3",
                            color=color))

ax.annotate('', xy=(3.15, 0.0), xycoords='data',
            xytext=(3.45, 0.45), textcoords='data',
            weight='bold', color=color,
            arrowprops=dict(arrowstyle='->',
                            connectionstyle="arc3",
                            color=color))

ax.text(4.0, -0.4, "Made with http://matplotlib.org",
        fontsize=10, ha="right", color='.5')

plt.show()
```
